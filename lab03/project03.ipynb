```python
import pandas as pd
import numpy as np
import json
import csv
```

# 1

```python
dfs = []
for file in ('data/proj3_data1.json','data/proj3_data2.json','data/proj3_data3.json'):
    data = pd.read_json(file)
    dfs.append(data)

df = pd.concat(dfs, ignore_index=True)
df.to_json('proj3_ex01_all_data.json')
df
```

# 2

```python
nans = []
for col in df.columns:
    sum = df[col].isna().sum()
    if sum > 0:
        nans.append([col, sum])
print(nans)

with open('proj3_ex02_no_nulls.csv', 'w') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerows(nans)
```

# 3

```python
with open('data/proj3_params.json') as json_file:
    json_params = json.load(json_file)
columns = json_params['concat_columns']

new_column = []
for index, row in df.iterrows():
    text = ""
    for col in json_params['concat_columns']:
        text += row[col]
        text += " "
    text = "".join([text[i] for i in range(len(text)-1)])
    new_column.append(text)

df['description'] = new_column
df.to_json('proj3_ex03_descriptions.json')
json_params
```

# 4

```python
df1 = pd.read_json('data/proj3_more_data.json')
join_column = json_params['join_column']
df_joined = pd.merge(df, df1, on=join_column, how='left')
df_joined.to_json('proj3_ex04_joined.json')
df_joined
```

# 5

```python
for i in df_joined.index:
    row = df_joined.loc[i]
    name = row['description'].lower().replace(' ','_')
    row = row.replace(np.nan, None)
    row = row.drop(['description']).to_dict()
    # df_joined.loc[i].to_json(f'ex05_{name}.json')
    with open(f'proj3_ex05_{name}.json', 'w') as outfile:
        json.dump(row, outfile)
for i in df_joined.index:
    row = df_joined.loc[i]
    row = row.replace(np.nan, None)
    name = row['description'].lower().replace(' ','_')
    row = row.drop(['description']).to_dict()
    for c in json_params['int_columns']:
        if pd.notnull(row[c]):
            row[c] = int(row[c])
        else:
            row[c] = None
        
    with open(f'proj3_ex05_int_{name}.json', 'w') as outfile:
        json.dump(row, outfile)
```

# 6

```python
print(json_params['aggregations'])
dict = {}
for column, aggregation in json_params['aggregations']:
    dict[f'{aggregation}_{column}'] = df_joined[column].agg(aggregation)
print(dict)
with open("proj3_ex06_aggregations.json", "w") as outfile: 
    json.dump(dict, outfile)
```

# 7

```python
df_grouped = df_joined.select_dtypes(include='number')
df_grouped = df_grouped.groupby(df_joined[json_params['grouping_column']]).mean()
df_grouped = df_grouped[df_joined.groupby(df[json_params['grouping_column']]).size() > 1]
df_grouped.to_csv('proj3_ex07_groups.csv', index=True, header=True)
df_grouped
```

# 8

```python
pivot_index = json_params['pivot_index']
pivot_columns = json_params['pivot_columns']
pivot_values = json_params['pivot_values']
print(pivot_index, pivot_columns, pivot_values)

df_pivot = df_joined.groupby([pivot_index, pivot_columns])[pivot_values].max()
df_pivot = df_pivot.reset_index().pivot(index=pivot_index, columns=pivot_columns, values=pivot_values)
df_pivot = df_pivot[df_pivot.notnull().any(axis=1)]

df_pivot.to_pickle('proj3_ex08_pivot.pkl')
df_pivot
```

```python
id_vars = json_params['id_vars']
df_melt = pd.melt(df_joined, id_vars=id_vars)

df_melt.to_csv('proj3_ex08_melt.csv', index=False, header=True)
df_melt
```

```python
wide_df = pd.read_csv('proj3_statistics.csv')
print(wide_df)
grouping_var = wide_df.columns[0]
wide_df = pd.wide_to_long(wide_df, 
                          wide_df.columns[1:].str.extract(r'^([^_]+)_', expand=False).unique(), 
                          i = grouping_var, 
                          j = 'Prefix', 
                          sep = '_')
wide_df.to_pickle('proj3_ex08_stats.pkl')
wide_df
```
